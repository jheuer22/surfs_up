```python
# Dependencies
import numpy as np
import pandas as pd

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func

#Addiotnal imports for analysis
import matplotlib.pyplot as plt
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract


# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_temps = session.query(Measurement.date, Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
print(june_temps)

```

```python
#  3. Convert the June temperatures to a list.
june_temps = []

# results = session.query(Measurement.date, Measurement.tobs)
# print(results.all())

june_temps = session.query(Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
print(june_temps)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temps_df = pd.DataFrame(june_temps, columns=['temperature'])

june_temps_df

# df.plot()
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temps_df.describe()
```

```python
#Additional Analysis for June 

june_temps_df = pd.DataFrame(june_temps, columns=['tobs'])
june_temps_df.plot.hist(bins=15)
plt.xlabel('Daily Temperatures')
plt.tight_layout()
plt.title('June Temperatures')
plt.xlim([60, 90])
plt.ylim([0,400])

# can see the average is around 75

# Save the figure.
plt.savefig("analysis/JuneTempsHist.png")
```

```python
# Create a box-and-whisker plot for the June temps.
x_labels = ["June"]
fig, ax = plt.subplots()
ax.boxplot(june_temps_df, labels=x_labels)
# Add the title, y-axis label and grid.
ax.set_title('June Temperatures')
ax.set_ylabel('Daily Temperatures')
ax.set_yticks(np.arange(62, 88, step=2.0))
ax.grid()

plt.tight_layout()
plt.savefig("analysis/JuneTempBoxPlot.png")
plt.show()
```
